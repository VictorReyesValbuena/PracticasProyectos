```python
"""
Empiezo creando la función make_constant_matrix, según las
instrucciones
"""
def make_constant_matrix(rows, cols, value):
    matrix = []
    for r in range(rows):
        row = []
        for c in range(cols):
            row.append(value)
        matrix.append(row)
    return matrix
```

```python
import random
def make_random_matrix(rows, cols):
    matrix = []
    for r in range(rows):
        row = []
        for c in range(cols):
            row.append(random.random()) # random.random() devuelve un float entre 0.0 y 1.0
        matrix.append(row)
    return matrix
```

```python
def get_valid_neighbors(i, j, max_rows, max_cols):
    """
    Recibe la posición (i, j) y el tamaño de la matriz.
    Devuelve una lista de coordenadas (fila, columna) de los vecinos válidos.
    """
    posibles_vecinos = [
        (i - 1, j), # Vecino superior
        (i + 1, j), # Vecino inferior
        (i, j - 1), # Vecino de la izquierda
        (i, j + 1)  # Vecino de la derecha
    ]
    vecinos_OK = []
    # Tenemos que asegurarnos de que solo nos quedamos con las coordenadas válidas, que estén dentro de la matriz.
    for fila, col in posibles_vecinos:
        # Verificamos que la fila sea válida (entre 0 y max_rows - 1)
        # Verificamos que la columna sea válida (entre 0 y max_cols - 1)
        if (0 <= fila < max_rows) and (0 <= col < max_cols):
            vecinos_OK.append((fila, col))
    return vecinos_OK
```

```python
def average_matrix(matrix):
    """
    Recibe una matriz y devuelve una nueva con los valores promediados.
    """
    total_rows = len(matrix)
    if total_rows > 0:
        total_cols = len(matrix[0])
    else:
        total_cols = 0
    new_matrix = [] # Aquí guardaremos el resultado final
    """
    'i' será el índice de la fila, 'row' es la lista de esa fila
    """
    for i, row in enumerate(matrix):
        new_row = [] # Fila nueva que estamos construyendo
        # 'j' será el índice de la columna, 'value' es el valor actual
        for j, value in enumerate(row):
            vecinos = get_valid_neighbors(i, j, total_rows, total_cols)
            #Con esto, obtengo los vecinos usando la función creada en el punto anterior.
            suma_total = value
            cuenta = 1
            for fila_vecino, col_vecino in vecinos:
                valor_vecino = matrix[fila_vecino][col_vecino]
                suma_total += valor_vecino
                cuenta += 1
            #Con esto sumo los valores de los vecinos encontrados.
            promedio = suma_total / cuenta
            new_row.append(promedio)
            #Con esto, calcúlo el promedio y lo añado a new_row
        new_matrix.append(new_row)
        #Este resultado, lo voy añadiendo a la variable new_matrix
    return new_matrix
```
