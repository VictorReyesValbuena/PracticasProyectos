## 1. Suma de los múltiplos

Crea la función `sum_mult_of_3`:

* Recibe una lista de números
* Devuelve la suma de aquellos que son **múltiplos de 3**

```python
def sum_mult_of_3(list_org):
    list_ok = []
    for i in list_org:
        if i % 3 == 0:
            list_ok.append(i)
    return list_ok
```

### 2. Suma de todos los elementos pares

Crea la función `sum_even` que recibe una lista de números y devuelve la suma de los que son pares.

```python
def sum_even (list_2):
    sum_par = 0
    for i in list_2:
        if i % 2 == 0:
            sum_par += i
    return sum_par
```

```python
sum_even([1,2,3,4,6])
```

### 3. Sumar números y None

Crea la función `flat_sum` que recibe una lista de *números opcionales* y devuelve el total.

Debes de recorrer todos los elementos, sumando los números e ignorando a los `None`s.

```python
def flat_sum(list_3):
    sum_all = 0
    for i in list_3:
        # Dentro del bucle for, pongo la condición por la cual,
        # solo sume aquellos valores de la lista que sean int
        if type(i) == int:
            sum_all += i
    return sum_all
```

```python
flat_sum([2, 3, 6, 3, None, 4])
```

Es decir, si recibe la lista `[1,2,3,4, 5, 6]`, el valor correcto de retorno es `[2,4,6]`. No es correcto devolver `[4, 2, 6]`.

## 6 - Crea la función  `select_contains`. Recibe dos parámetros:

1. Una cadena de cualquier longitud
2. Una lista de cadenas

Devuelve una lista de aquellas cadenas que contienen la recibida como parámetro, en cualquier posición (principio, em medio, final, no importa)

```python
def select_contains(cadena, list_contains):
    list_contains_ok = []
    for i in list_contains:
        if len(cadena) <= len(i):
            if cadena in i:
                list_contains_ok.append(i)
    return list_contains_ok
```

## Construye una función llamada `make_lengths` que recibe una lista de cadenas y devuelve: una *lista de números* , de la *misma longitud* de la lista de entrada. Cada uno de esos números es la longitud de cada una de las cadenas de entrada.

Por ejemplo:

`["hey", "hola", "mundo"]` --> `[3, 4, 5]`
`[]` --> `[]`

```python
def make_lengths(chain_list):
    list_number = []
    for i in chain_list:
        list_number.append(len(i))
    return list_number
```

```python
make_lengths(["hey", "hola", "mundo"])
```

## Crea la función `update_salaries` que recibe una lista de nóminas y crea una nueva lista con las nóminas aplicando los siguientes cambios:

* nómina < 900 se incrementa en un 30%
* 900 <= nómina < 1500, se incrementa en un 12%
* 1500 <= nómina < 6000, se queda tal cual
* nómina > 6000, se incrementa en un 100% (ventajas de ser alto cargo)

```python
def update_salaries(nominas):
    nominas_mod = []
    for i in nominas:
        if i < 900:
            nominas_mod.append((i*1.30))
        if 900 <= i < 1500:
            nominas_mod.append((i*1.12))
        if 1500 <= i < 6000:
            nominas_mod.append(i)
        if i > 6000:
            nominas_mod.append((i*2))
    return nominas_mod
```

```python
update_salaries([4000,2000,4500,7000,500])
```

## Crea la función `explode`. Recibe una cadena y la convierte en una lista de caracteres.

Por ejemplo:

`"hola" --> ["h", "o", "l", "a"]`

`"" --> []`

```python
def explode(cadena):
    list_cadena = []
    for i in cadena:
        list_cadena.append(i)
    return list_cadena
```

## La *Lengua de la P* es un juego de niños común en Brasil en el que se alteran las palabras convirtiéndolas en una especie de trabalenguas.

Si una letra es una vocal, se le transforma en una sílaba con la p y las dos vocales:

* a -> apa
* e -> epe
* i -> ipi
* o -> opo
* u -> upu

Aplicando esto, las siguientes palabras se convertirían así:

* Basico -> Bapasipicopo
* Raro -> Raparopo
* Alamo -> Apalapamopo

Crea la función `to_p_sentence` que recibe una frase (una cadena) y la convierte a la *Lengua de la P*.

```python
def to_p_sentence(frase):
    Lengua_de_la_P = ""
    vocales = "aeiou"
    for i in frase:
        if i not in vocales:
            Lengua_de_la_P += i
        if i in vocales:
            Lengua_de_la_P += i+"p"+i
    return Lengua_de_la_P
```

```python
to_p_sentence("Raro")
```

# Aparece un nuevo cliente y te dice que quiere algo similar, pero la para la *"f"*. Te pide que implementes la función `dinofauriza(text)`. Lo único que cambia es que en vez de la *p* usarías la *f*. **No vas a escribir una función sólo para eso**.

Así que decides escribir la función `to_consonant(text, consonant)` que recibe una cadena a ser procesada y una consonante que se usará para los cambios.

1. Crea la función `to_consonant`
2. Refactoriza la función `to_p_sentence` para que sólo sea una llamada a `to_consonant`.
3. Crea la función `dinofauriza` que lo único que hace es llamar a `to_consonant`
4. ¡Comprueba que funciona! Usa `assert`

```python
def to_p_sentence(frase,consonante):
    Lengua_de_la_P = ""
    vocales = "aeiou"
    for i in frase:
        if i not in vocales:
            Lengua_de_la_P += i
        if i in vocales:
            Lengua_de_la_P += i+consonante+i
    return Lengua_de_la_P
```

```python
def to_consonant(text, consonant):
    return to_p_sentence(text, consonant)
```

```python
to_consonant("Raro", "l")
```

```python
def dinofauriza(palabra):
    return to_consonant(palabra, "f")
```
