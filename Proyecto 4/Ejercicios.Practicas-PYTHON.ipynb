Crea una función que devuelva una lista con los número pares que se encuentren dentro de una secuencia. Se debe indicar el número de inicio y número de final

```python
def pares (x,y):
    list = []
    for i in range(x,y+1):
        if i % 2 == 0:
           list.append(i)
    return list
```

```python
pares (10,20)
```
